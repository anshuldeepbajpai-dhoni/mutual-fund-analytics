### Imports

```python
import pandas as pd
import numpy as np
import sqlite3

from pathlib import Path
from sqlalchemy import create_engine
```

### Load All Datasets

```python
raw_path = "../data/raw"

fund_master = pd.read_csv(
    f"{raw_path}/01_fund_master.csv"
)

nav_history = pd.read_csv(
    f"{raw_path}/02_nav_history.csv"
)

aum = pd.read_csv(
    f"{raw_path}/03_aum_by_fund_house.csv"
)

sip = pd.read_csv(
    f"{raw_path}/04_monthly_sip_inflows.csv"
)

category = pd.read_csv(
    f"{raw_path}/05_category_inflows.csv"
)

folio = pd.read_csv(
    f"{raw_path}/06_industry_folio_count.csv"
)

performance = pd.read_csv(
    f"{raw_path}/07_scheme_performance.csv"
)

transactions = pd.read_csv(
    f"{raw_path}/08_investor_transactions.csv"
)

holdings = pd.read_csv(
    f"{raw_path}/09_portfolio_holdings.csv"
)

benchmark = pd.read_csv(
    f"{raw_path}/10_benchmark_indices.csv"
)
```

### Dataset Inspection 

```python
datasets = {
    "fund_master": fund_master,
    "nav_history": nav_history,
    "aum": aum,
    "sip": sip,
    "category": category,
    "folio": folio,
    "performance": performance,
    "transactions": transactions,
    "holdings": holdings,
    "benchmark": benchmark
}

for name, df in datasets.items():

    print("="*60)
    print(name.upper())
    print("="*60)

    print("Shape:", df.shape)
    print(df.dtypes)

    display(df.head())
```

### Clean NAV History

```python
nav_history["date"] = pd.to_datetime(
    nav_history["date"]
)

nav_history = nav_history.sort_values(
    ["amfi_code","date"]
)

nav_history["nav"] = (
    nav_history
    .groupby("amfi_code")["nav"]
    .ffill()
)

nav_history = nav_history.drop_duplicates()

nav_history = nav_history[
    nav_history["nav"] > 0
]
```

### Clean Transactions

```python
transactions["transaction_date"] = pd.to_datetime(
    transactions["transaction_date"]
)

transactions = transactions[
    transactions["amount_inr"] > 0
]

transactions["transaction_type"] = (
    transactions["transaction_type"]
    .str.strip()
    .str.title()
)
```

### Clean Performance Data 

```python
performance["return_1yr_pct"] = pd.to_numeric(
    performance["return_1yr_pct"],
    errors="coerce"
)

performance["return_3yr_pct"] = pd.to_numeric(
    performance["return_3yr_pct"],
    errors="coerce"
)

performance["return_5yr_pct"] = pd.to_numeric(
    performance["return_5yr_pct"],
    errors="coerce"
)
```

### Save Clean Files

```python
processed = "data/processed"

Path(processed).mkdir(
    parents=True,
    exist_ok=True
)

nav_history.to_csv(
    f"{processed}/02_nav_history_clean.csv",
    index=False
)

transactions.to_csv(
    f"{processed}/08_investor_transactions_clean.csv",
    index=False
)

performance.to_csv(
    f"{processed}/07_scheme_performance_clean.csv",
    index=False
)
```

### Create SQLite Database

```python
engine = create_engine(
    "sqlite:///../database/bluestock_mf.db"
)

print("Database Connected")
```

### Load Data into SQLite 

```python
fund_master.to_sql(
    "fund_master",
    engine,
    if_exists="replace",
    index=False
)

nav_history.to_sql(
    "nav_history",
    engine,
    if_exists="replace",
    index=False
)

performance.to_sql(
    "scheme_performance",
    engine,
    if_exists="replace",
    index=False
)

transactions.to_sql(
    "investor_transactions",
    engine,
    if_exists="replace",
    index=False
)
```

### Verify Row Counts

```python
conn = sqlite3.connect(
    "../database/bluestock_mf.db"
)

tables = [
    "fund_master",
    "nav_history",
    "scheme_performance",
    "investor_transactions"
]

for table in tables:

    query = f"""
    SELECT COUNT(*)
    FROM {table}
    """

    count = pd.read_sql(
        query,
        conn
    )

    print(table)
    print(count)
```
